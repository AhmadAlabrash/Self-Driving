```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

```

```python
image = cv2.imread('/content/stop-sign1.jpg')
plt.imshow(image)
```

```python
plt.plot(100,350,'+')
plt.plot(465,420,'+')
plt.plot(100,120,'+')
plt.plot(465,240,'+')
plt.imshow(image)
```

```python
src =np.float32(
    [
     [100,350],
     [465,420],
     [100,120],
     [465,240]
    ]
)

dst =np.float32(
    [
     [140,350],
     [485,400],
     [140,120],
     [485,220]
    ]
)
```

```python
m = cv2.getPerspectiveTransform(src,dst)
```

```python
wrapped_image = cv2.warpPerspective(image,m,(image.shape[1],image.shape[0]),cv2.INTER_LINEAR)
```

```python
plt.imshow(wrapped_image)
```
